```python


# Calculate mean, median, and total number of cells
mean_conf_score = cell_metadata['final_level_softmax_prob'].mean()
median_conf_score = cell_metadata['final_level_softmax_prob'].median()
total_cells = cell_metadata['final_level_softmax_prob'].shape[0]

# Plot histogram
cell_metadata['final_level_softmax_prob'].hist(bins=50, figsize=(10, 6), alpha=0.7, label='Confidence Scores')
plt.axvline(mean_conf_score, color='red', linestyle='dashed', linewidth=1.5, label=f'Mean: {mean_conf_score:.6f}')
plt.axvline(median_conf_score, color='blue', linestyle='dashed', linewidth=1.5, label=f'Median: {median_conf_score:.6f}')
plt.title('Distribution of Softmax Scores')
plt.xlabel('final lvl Softmax Score')
plt.ylabel('Frequency')
plt.legend()

# Display summary metrics
num_scores_above_90 = (cell_metadata['final_level_softmax_prob'] > 0.9).sum()
summary_text = (
    f"Min: {cell_metadata['final_level_softmax_prob'].min():.6f}\n"
    f"Max: {cell_metadata['final_level_softmax_prob'].max():.6f}\n"
    f"Total Cells: {total_cells}\n"
    f"Scores > 90%: {num_scores_above_90}"
)
plt.figtext(0.2, 0.4, summary_text, fontsize=10, bbox={"facecolor": "white", "alpha": 0.5, "pad": 5})

plt.show()
```

```python
import json

# Create mapping from Azimuth cell types to simplified categories
azimuth_to_simplified = {
    # Macrophages
    'Alveolar macrophage': 'Macro',
    'Macrophage': 'Macro', 
    'M1 macrophage': 'Macro',
    'LYVE1 macrophage': 'Macro',
    
    # Monocytes
    'CD14 monocyte': 'Monocyte',
    'Monocyte': 'Monocyte',
    'CD16 monocyte': 'Monocyte',
    
    # T cells - CD4
    'Memory CD4 T cell': 'CD4',
    'CD4 T cell': 'CD4',
    'Naive CD4 T cell': 'CD4',
    'Treg cell': 'CD4',  # Regulatory T cells are CD4+
    'Tfh cell': 'CD4',  # T follicular helper cells are CD4+
    
    # T cells - CD8
    'GZMK CD8 T cell': 'CD8',
    'CD8 T cell': 'CD8',
    'GZMB CD8 T cell': 'CD8',
    'GZMK+IL7R+ CD8 T cell': 'CD8',
    'Memory CD8 T cell': 'CD8',
    'GZMK+IL7R- CD8 T cell': 'CD8',
    
    # T cells - General
    'T cell': 'T',  # General T cell category
    'Cycling T/NK cell': 'T',  # Cycling cells, predominantly T
    
    # NK cells
    'CD16 NK cell': 'NK',
    'NK cell': 'NK',
    
    # B cells and Plasma cells
    'Plasma cell': 'B',  # Plasma cells are differentiated B cells
    'Naive B cell': 'B',
    'B cell': 'B',
    'Memory B cell': 'B',
    
    # Dendritic cells
    'cDC2': 'DC',
    'cDC': 'DC',  # Conventional dendritic cells
    
    # Mast cells
    'Mast cell': 'Mast',
    
    # Epithelial cells - Alveolar
    'AT2': 'AT2',
    'AT1': 'AT1',
    
    # Epithelial cells - Secretory
    'Club cell': 'Secretory',  # Club cells are secretory epithelial cells
    'Goblet cell': 'Secretory',  # Goblet cells are secretory
    'TFF1+ ductal cell': 'Secretory',  # Ductal cells are often secretory
    
    # Epithelial cells - Other
    'CLDN3+ epithelial cell': 'other',  # Generic epithelial
    
    # Ciliated cells
    'Multiciliated cell': 'Ciliated',
    
    # Endothelial cells
    'Venous EC': 'Endothelia',
    'Capillary EC': 'Endothelia',
    
    # Fibroblasts
    'IGFBP6+APOD+ fibroblast': 'Fibro',
    
    # Other/Proliferating cells
    'S phase myeloid cell': 'other'  # Proliferating myeloid cells
}

# Add default mapping for any other cell types
def map_cell_type(cell_type):
    return azimuth_to_simplified.get(cell_type, 'other')

# Test the mapping on known cell types
print("Mapping preview:")
for azimuth_type, simplified_type in list(azimuth_to_simplified.items())[:10]:
    print(f"  {azimuth_type} -> {simplified_type}")

print(f"\nTotal mapped cell types: {len(azimuth_to_simplified)}")
print("Any unmapped cell types will be assigned to 'other'")

# Save the mapping to JSON file
output_path = './azimuth_to_simplified_mapping.json'
with open(output_path, 'w') as f:
    json.dump(azimuth_to_simplified, f, indent=2)

print(f"\nMapping saved to: {output_path}")
```

```python
# Apply the mapping to create simplified cell type categories
cell_metadata['simplified_cell_type'] = cell_metadata['final_level_labels'].apply(map_cell_type)

# Create simplified cell type counts
simplified_counts = cell_metadata['simplified_cell_type'].value_counts().reset_index()
simplified_counts.columns = ["Cell Type", "Count"]

# Calculate proportions
simplified_counts['Proportion'] = simplified_counts['Count'] / simplified_counts['Count'].sum()

# Save simplified proportions to CSV
simplified_output_path = './azimuth_simplified_proportions.csv'
simplified_counts.to_csv(simplified_output_path, index=False)
print(f"\nSimplified proportions saved to: {simplified_output_path}")
```

```python
# Save proportions per cell type with cell types as row names and 'SC' as the column name
azimuth_proportions = simplified_counts.set_index("Cell Type")[["Proportion"]].rename(columns={"Proportion": "SC"})
azimuth_proportions.to_csv("azimuth_proportions.csv")
print("Proportions saved to azimuth_proportions.csv")
```
