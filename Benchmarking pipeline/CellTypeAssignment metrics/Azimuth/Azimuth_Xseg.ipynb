```python
import panhumanpy as ph
import matplotlib.pyplot as plt
import scanpy as sc
import pandas as pd
from matplotlib.patches import Patch
import json
import numpy as np
```

```python

cells_adata_path = "/srv/gstore/projects/p37785/EzPyzENACTApp_2025-07-07--09-51-04/ENACTApp_Visium_HD_Human_Lung_Cancer_w_Xenium_seg/chunks/weighted_by_area/celltypist_results/cells_adata.h5"

method_name = "Xseg"
```

```python
adata = sc.read_h5ad(cells_adata_path)
```

```python
# High-level interface
azimuth = ph.AzimuthNN(adata)
cell_metadata = azimuth.cells_meta
embeddings = azimuth.azimuth_embed()
umap = azimuth.azimuth_umap()
```

```python
# Calculate mean, median, and total number of cells
mean_conf_score = cell_metadata['final_level_softmax_prob'].mean()
median_conf_score = cell_metadata['final_level_softmax_prob'].median()
total_cells = cell_metadata['final_level_softmax_prob'].shape[0]

# Plot histogram
cell_metadata['final_level_softmax_prob'].hist(bins=50, figsize=(10, 6), alpha=0.7, label='Confidence Scores')
plt.axvline(mean_conf_score, color='red', linestyle='dashed', linewidth=1.5, label=f'Mean: {mean_conf_score:.6f}')
plt.axvline(median_conf_score, color='blue', linestyle='dashed', linewidth=1.5, label=f'Median: {median_conf_score:.6f}')
plt.title('Distribution of Softmax Scores')
plt.xlabel('final lvl Softmax Score')
plt.ylabel('Frequency')
plt.legend()

# Display summary metrics
num_scores_above_90 = (cell_metadata['final_level_softmax_prob'] > 0.9).sum()
summary_text = (
    f"Min: {cell_metadata['final_level_softmax_prob'].min():.6f}\n"
    f"Max: {cell_metadata['final_level_softmax_prob'].max():.6f}\n"
    f"Total Cells: {total_cells}\n"
    f"Scores > 90%: {num_scores_above_90}"
)
plt.figtext(0.2, 0.4, summary_text, fontsize=10, bbox={"facecolor": "white", "alpha": 0.5, "pad": 5})

plt.show()
```

```python
import json

# Load mapping from Azimuth cell types to simplified categories
with open('azimuth_to_simplified_mapping.json', 'r') as f:
    azimuth_to_simplified = json.load(f)

def map_cell_type(cell_type):
    return azimuth_to_simplified.get(cell_type, 'other')

print("Mapping preview:")
for azimuth_type, simplified_type in list(azimuth_to_simplified.items())[:10]:
    print(f"  {azimuth_type} -> {simplified_type}")

print(f"\nTotal mapped cell types: {len(azimuth_to_simplified)}")
print("Any unmapped cell types will be assigned to 'other'")
```

```python
# Apply the mapping to create simplified cell type categories
cell_metadata['simplified_cell_type'] = cell_metadata['final_level_labels'].apply(map_cell_type)
adata.obs['simplified_cell_type'] = cell_metadata['simplified_cell_type']
# Create simplified cell type counts
simplified_counts = cell_metadata['simplified_cell_type'].value_counts().reset_index()
simplified_counts.columns = ["Cell Type", "Count"]

# Calculate proportions
simplified_counts['Proportion'] = simplified_counts['Count'] / simplified_counts['Count'].sum()
```

```python
# Load the CSV file
azimuth_prop_df = pd.read_csv('azimuth_proportions.csv')

# Append the computed proportions from simplified_counts
azimuth_prop_df = azimuth_prop_df.merge(
    simplified_counts[['Cell Type', 'Proportion']],
    on='Cell Type',
    how='left'
)
azimuth_prop_df = azimuth_prop_df.rename(columns={'Proportion': method_name})

# Save the updated dataframe
azimuth_prop_df.to_csv('azimuth_proportions.csv', index=False)

azimuth_prop_df.head()
```

```python
label_list = ["AT1", "AT2", "B","Basal", "CD4", "CD8","Ciliated","DC", "Endothelia", "Fibro",
"Macro","Mast","Monocyte", "NKT","NK","Secretory", "T" , "other"] 

colors = plt.cm.tab20.colors  # Use a colormap with enough distinct colors
cell_types = np.sort(label_list)

# Assign a color to each cell type
color_map = {cell_type: colors[i % len(colors)] for i, cell_type in enumerate(cell_types)}
# Plot the spatial scatter as before
plt.figure(figsize=(12, 10))
plt.scatter(
    adata.obsm['spatial']['cell_x'],
    adata.obsm['spatial']['cell_y'],
    s=2,
    alpha=0.5,
    c=[color_map[cell_type] for cell_type in adata.obs["simplified_cell_type"]],
    label='Cells'
)

plt.title('Dotplot of Cells colored by Broad Cell Type')
plt.xlabel('X Coordinate')
plt.ylabel('Y Coordinate')

# Create custom legend
legend_elements = [
    Patch(facecolor=color_map[cell_type], label=cell_type)
    for cell_type in cell_types
]
plt.legend(handles=legend_elements, title="Broad Cell Type", bbox_to_anchor=(1.05, 1), loc='upper left', borderaxespad=0.)

plt.tight_layout()
plt.show()
```
